```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
sns.set_style("darkgrid")
os.getcwd()
```

```python
iris_train = pd.read_csv('../data/prepared/train.csv', sep=',', header=0)
iris_test  = pd.read_csv('../data/prepared/test.csv',  sep=',', header=0)
```

```python
iris = pd.concat([iris_train, iris_test], ignore_index=True).sort_values(by=['target']).reset_index(drop=True)
display(iris.head())
```

```python
classes = {
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
}

iris.target = [classes[i] for i in iris.target]
display(iris.head())
```

```python
iris.info()
```

```python
fig = iris[iris.target=='setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
iris[iris.target=='versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='versicolor',ax=fig)
iris[iris.target=='virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
fig = iris[iris.target=='setosa'].plot(kind='scatter',x='petal_length',y='petal_width',color='orange', label='Setosa')
iris[iris.target=='versicolor'].plot(kind='scatter',x='petal_length',y='petal_width',color='blue', label='versicolor',ax=fig)
iris[iris.target=='virginica'].plot(kind='scatter',x='petal_length',y='petal_width',color='green', label='virginica', ax=fig)
fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title("Petal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

```python
iris.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='target',y='petal_length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='target',y='petal_width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='target',y='sepal_length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='target',y='sepal_width',data=iris)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, square=True)

heatmap.set_title('Correlation Heatmap', fontsize=16)

heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=8)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=8)

plt.show()
```
